```python
import pandas 
import statsmodels as sm
import matplotlib.pyplot as plt 
```

```python
data = sm.datasets.co2.load_pandas()
co2 = data.data

print(co2.shape)
co2.head()
```

```python
co2_monthly = co2.resample('ME').mean().ffill()
co2_monthly.head()
```

```python
from statsmodels.tsa.seasonal import STL

res = STL(co2_monthly).fit()
res.plot()
plt.show()
```
