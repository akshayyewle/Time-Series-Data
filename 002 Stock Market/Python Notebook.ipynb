```python
# Plot Closing Prices
Stock_Data.groupby([('Ticker')])['Close']
```

# **Ford Data**

```python
Ford = yf.Ticker('F').history(period="max",interval='1d')
Ford.head()
```

```python
Ford['Daily_Return'] = Ford['Close'] - Ford['Close'].shift(1) 
Ford['Daily_Return_Pct'] = Ford['Close'].pct_change() * 100
Ford.head()
```

```python
Ford['Daily_Return_Pct'].plot(kind='hist',bins=150,figsize=(6,4))
plt.xlim(-10,10)
plt.title('Daily Return %')
plt.show()
```

# **S&P500 Data**

```python
SP500 = yf.Ticker('^GSPC').history(period="max",interval='1d')
SP500.head()
```

```python
plt.plot(SP500.index, SP500['Close'])
plt.title('SP500')
plt.show()
```

# **Decomposition**

```python
import statsmodels.api as sm

sm.tsa.seasonal_decompose(x=SP500['Close'],model='additive')
```
