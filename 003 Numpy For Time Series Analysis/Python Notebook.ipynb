# **Setup**

```python
import numpy as np
import matplotlib.pyplot as plt
```

# **Data Import**

```python
dates = np.array(['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'], dtype='datetime64')
dates
```

## **Create A Date Range**

```python
date_range = np.arange('2023-01-01', '2025-01-01', dtype='datetime64[M]')

print(date_range.shape)
display(date_range)
```

```python
data = np.random.randn(len(date_range)) * 10 + 100 

print(data.shape)
display(data)
```

```python
plt.figure(figsize=(10,4))

plt.plot(date_range,data,linestyle='solid')

# Adding titles and labels
plt.title('Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Value')

# Display the plot
plt.show()
```

# **Moving Average**

```python
def moving_average(data, window):
    return np.convolve(data, np.ones(window), 'valid') / window

ma_12 = moving_average(data, 12)

print(ma_12.shape)
display(ma_12)

# Plot the moving averages
plt.figure(figsize=(10,4))
plt.plot(date_range[:13],ma_12,linestyle='solid')
plt.title('Moving Averages Plot')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
```
