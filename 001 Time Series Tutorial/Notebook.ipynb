# **Setup**

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.tsa.seasonal import seasonal_decompose
```

# **Data Import**

```python
df = pd.read_csv('AirPassengers.csv', parse_dates=['Month'])

print(df.shape)
print(df.columns)

df.head()
```

```python
# Set Month Column As Index
df = df.set_index('Month')
df.head()
```

```python
df.asfreq('ME')
df.head()
```

# **Plot TS**

```python
df.plot(y='#Passengers')
```

# **TS Decomposition** 

```python
seasonal_decompose(df['#Passengers'], model='additive', period=12).plot()
```
